**Overview**

This is a simple Tkinter GUI application that displays a help message when a button is clicked. The message provides basic usage instructions for MySQL commands.
Code Breakdown

    Importing Required Modules

import tkinter as tk

    tkinter is used to create the graphical user interface.

    Creating the Main Window

root = tk.Tk()
root.geometry('600x400')
root.resizable(True, True)

    Tk() initializes the main application window.
    geometry('600x400') sets the window size to 600x400 pixels.
    resizable(False, False) allows resizing of the window.

    Creating the Label Widget

label = tk.Label(root, text="0")
label.pack(pady=5)

- Label(root, text="0") creates a label widget with an initial value of "0".
- pack(pady=5) places the label in the window with padding.

- justify="left" aligns the text to the left.
- font=("calibri", 10) sets the font style and size.
- pack(pady=10) adds padding for spacing.

- Defining the Help Message Function
```
def help_message():
    message = '''
    Usage: mysql command
    -h hostname     
    -d database name
    -u username
    -p password 
    '''
```
- label.config(text=message)
- help_message() is a function that defines a multi-line help message.
- label.config(text=message) updates the label to display the message when the function is triggered.

- Creating the Button Widget
```
btn = tk.Button(root, text="Show message", command=help_message)
btn.pack(pady=5)
```
- Button(root, text="Show message", command=help_message) creates a button labeled "Show message".
- Clicking the button calls help_message(), displaying the help text in the label.
- pack(pady=5) positions the button with spacing.


- Running the Application

`root.mainloop()`

- mainloop() starts the Tkinter event loop, keeping the GUI application running.
  
Expected Behavior

- The application starts with a blank label and a "Show message" button.
- Clicking the button displays a MySQL usage message inside the label.
- The window remains fixed in size (400x400 pixels).


````python
import tkinter as tk

root = tk.Tk()
root.geometry('400x400')
root.resizable(False,False)

label = tk.Label(root, text="", justify="left", font=("calibri", 10))
label.pack(pady=10)

def help_message():
    message = '''
    Usage: mysql command
    -h hostname     
    -d database name
    -u username
    -p password 
     '''
    label.config(text=message)
    
btn = tk.Button(root, text="Show message", command=help_message)
btn.pack(pady=5)

root.mainloop()

````
